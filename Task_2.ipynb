### importing the libraries


```python
import pandas as pd 
import matplotlib.pyplot as plt
```

### Loading the data set and understanding 

```python
#importing data 
data = pd.read_csv("StudentsPerformance.csv")
data
```

## Data Validation ( is their any null values  or any negative marks are their)

```python
data.isnull().values.any()
```

```python
# Check for negative marks
if (data[['math score','reading score','writing score']] < 0).any().any():
    print("Dataset contains negative marks. Please correct them.")
    data = data[data[['math score', 'reading score', 'writing score']].ge(0).all(axis=1)]  # Keep only non-negative rows
else:
    print("their is no negative  marks given")
```

## Average  Marks Per Subjects 

```python
# average mark per math 
avg_marks=data[['math score','reading score','writing score']].mean()
print("average marks per subject is :")
avg_marks
```

## Identify and display top 5 performing students.

```python
data['total_score']=data['math score'] +data ['reading score'] + data['writing score']
top_students=data.sort_values(by="total_score",ascending=False).head(5)
top_students
```

## Visualize performance using bar or histogram charts.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
avg_marks.plot(kind='bar', color=['skyblue', 'red', 'yellow'])
plt.title('Average Marks per Subject')
plt.ylabel('Marks')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(data['math score'], bins=10)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
avg_gender = data.groupby('gender')[['math score','reading score','writing score']].mean()

plt.figure(figsize=(8,5))
avg_gender.plot(kind='bar')
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```

```python
parent_avg = data.groupby('parental level of education')[['math score','reading score','writing score']].mean()

plt.figure(figsize=(12,6))
parent_avg.plot(kind='bar')
plt.title('Parental Education vs Student Performance')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.show()
```

```python
race_avg = data.groupby('race/ethnicity')[['math score','reading score','writing score']].mean()

plt.figure(figsize=(10,6))
race_avg.plot(kind='bar')
plt.title('Race/Ethnicity vs Average Scores')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```

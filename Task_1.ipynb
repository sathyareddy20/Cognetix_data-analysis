## Importing the libraries

```python
# importing the required libraries 
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib. as plt
df.columns
```

## loading the dataset

```python
# importing the csv data using pandas
df=pd.read_csv("sales_data_sample.csv",encoding="latin1")
df
```

```python
# basic cleaning of a dataset 
df.isnull().sum()
df.drop_duplicates()
df.columns=df.columns.str.strip().str.lower()
df
```

## statistics of a sales column

```python

# these are the basic statistics of a data set 
totalsales=df['sales'].sum()
avg_sales=df['sales'].mean()
max_sales=df['sales'].max()
min_sales=df['sales'].min()
print("totalsales are:" ,totalsales)
print("averagesales are:",avg_sales)
print("maximum sales are:",max_sales)
print("minimun sales are :",min_sales)
```

## Bar and Pie chart for category wise 

```python
import matplotlib.pyplot as plt
dir(plt)
catogory_sales=df.groupby('productline')['sales'].sum().sort_values(ascending=False)
print(category_sales)
#Create bar and pie charts for category-wise or region-wise sales using Matplotlib/Seaborn.
category_sales.plot(kind='bar', figsize=(6,4))
plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140)
plt.title("Sales Distribution by Product Line")
plt.show()
```

## Region-wise bar and piechart

```python
df['country']
```

```python
region_sales=df.groupby('country')['sales'].sum()
plt.figure(figsize=(8,8))
plt.pie(region_sales, labels=region_sales.index,autopct='%1.1f%%',startangle=100)
plt.title("Sales Distribution by Region")
plt.show()
```

```python
regionwise_sales=df.groupby('country')['sales'].sum().sort_values(ascending=False)
regionwise_sales.plot(kind='bar', figsize=(6,4))
plt.title("regionwise sales ")
plt.xlabel("country/regions")
plt.ylabel("Total Sales")
plt.show()
```
